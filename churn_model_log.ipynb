```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Dosajh/Documents/Python Scripts/new_churn_data.csv')
```

```python
import random
```

```python
user = data['user']
data = data.drop(columns='user')
```

```python
data.housing.value_counts()
```

```python
data.groupby('housing')['churn'].nunique().reset_index()
```

```python
data = pd.get_dummies(data)
```

```python
data
```

```python
data = data.drop(columns = ['housing_na', 'zodiac_sign_na', 'payment_type_na'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns = 'churn'), data['churn'],
                                                    test_size = 0.2,
                                                    random_state = 0)
```

```python
pos_index = y_train[y_train.values == 1].index
neg_index = y_train[y_train.values == 0].index
if len(pos_index) > len(neg_index):
    higher = pos_index
    lower = neg_index
else:
    higher = neg_index
    lower = pos_index

random.seed(0)
higher = np.random.choice(higher, size=len(lower))
lower = np.asarray(lower)
new_indexes = np.concatenate((lower, higher))

X_train = X_train.loc[new_indexes,]
y_train = y_train[new_indexes]
```

```python
x_train=preprocessing.normalize(X_train)
x_test = preprocessing.normalize(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python

print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("logistic reg Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```

```python
pd.concat([pd.DataFrame(X_train.columns, columns = ["features"]),
           pd.DataFrame(np.transpose(classifier.coef_), columns = ["coef"])
           ],axis = 1)

#coefficients
```

```python
final_results = pd.concat([y_test, user], axis = 1).dropna()
final_results['predicted_churn'] = y_pred
final_results = final_results[['user', 'churn', 'predicted_churn']].reset_index(drop=True)
```
